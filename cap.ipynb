```python


def compare_dataframes(df_n, df_m):
    # Compare index
    index_match = df_n.index.equals(df_m.index)

    # Compare columns
    columns_match = df_n.columns.equals(df_m.columns)

    # Compare number of rows
    rows_match = len(df_n) == len(df_m)

    return index_match, rows_match, columns_match

index_match, rows_match, columns_match = compare_dataframes(df_n, df_m)

if index_match and rows_match and columns_match:
    print("The index, rows, and columns match between the two data frames.")
else:
    print("The index, rows, and/or columns do not match between the two data frames.")
```
